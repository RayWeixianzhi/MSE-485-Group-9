```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 30, 300)
y = []
y2 = []
y3 = []
```

```python
from matplotlib import axes
for line in open('C4-rep1-mean.txt', 'r'):
  values = []
  values = [float(s) for s in line.split()]
  y.append (values[0])

for line in open('C4_rep2_mean.txt', 'r'):
  values = []
  values = [float(s) for s in line.split()]
  y2.append (values[0])  

 
for line in open('C4_rep3_mean.txt', 'r'):
  values = []
  values = [float(s) for s in line.split()]
  y3.append (values[0])  

plt.figure(figsize=(8,3))
plt.plot(x, y, label = "C4_rep1")
plt.plot(x, y2, label = "C4_rep2")
plt.plot(x, y3, label = "C4_rep3")
plt.xlabel("Time (ns)")
plt.ylabel("Diffusion coefficient (cm^2/s)")
plt.ylim([0, 4e-6])
plt.legend()
#plt.show(dpi = 600)
plt.tight_layout()
plt.savefig("test.png",dpi=600)
```
